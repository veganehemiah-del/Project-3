```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("athlete_events(in).csv")
df=df.fillna("0") #filled with zero as analyzing height/weight was not a priority
df.head()
```

```python
df.columns.tolist() #to get confirmation on the exact names of the categories 
```

```python
amt_of_medals=df[df['Medal'].notna()].groupby('NOC')['Medal'].count().sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 8))
sns.histplot(x=amt_of_medals.index, y=amt_of_medals.values, color='lightblue', edgecolor='orange')
plt.xlabel('NOC')
plt.ylabel('Number of Medals')
plt.title(b'Top 15 Countries By Medal Count')
plt.show()
#as to seperate and confidently show the differences in values between countires through (NOC) label 
```

```python
amt_of_olympians_per_sex=df[df['Sex'].notna()].groupby('Sex')['ID'].count().sort_values(ascending=False).head(27116)
plt.figure(figsize=(12, 8))
sns.barplot(x=amt_of_olympians_per_sex.index, y=amt_of_olympians_per_sex.values, color='lightblue')
plt.xlabel('Sex')
plt.ylabel('ID')
plt.title('Amount of Men and Women Olympians')
plt.show()
#as to show the numerical, participant difference between the sexes
```
